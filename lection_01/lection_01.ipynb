## Spannungsanalyse
### Theorie

Ziel dieser Aufgabe ist die Eigenvektoren $\underline{v}_i$ und ihre korrespondierenden Eigenwerte $ \lambda_i $ (mit $ i=I,II $) eines zweideimensionalen Spannungszustandes zu berechnen. Ausgehend von einem allgemeinen Spannungszustand (2D) <br>

$$\underline{\underline{\sigma}}= \begin{bmatrix} \sigma_x & \tau_{xy} \\ \tau_{yx} & \sigma_y \end{bmatrix}$$ 

kann die charakteristische Gleichung wie folgt geschrieben werden



---
#### Eigenwert

$$\det{\left(\underline{\underline{\sigma}}-\lambda \underline{\underline{1}} \right)} \overset{!}{=} 0\qquad \text{mit}\; \underline{\underline{1}}=\begin{bmatrix} 1 & 0 \\ 0 & 1\end{bmatrix} \text{(Einheitsmatrix)}$$


$$\det{\left(\begin{bmatrix} \sigma_x & \tau_{xy} \\ \tau_{yx} & \sigma_y \end{bmatrix}-\lambda \begin{bmatrix} 1 & 0 \\ 0 & 1\end{bmatrix}\right)} = =$$

$$\det{\left(\begin{bmatrix} \sigma_x-\lambda & \tau_{xy} \\ \tau_{yx} & \sigma_y-\lambda \end{bmatrix}\right)} = 0 $$

Folglich führt das zu einer quadratischen Gleichung (im 2D-Fall, kubische Gleichung für den 3D-Fall)

$$ \left( \sigma_x-\lambda \right)\left( \sigma_y-\lambda \right)-\tau_{xy}\tau_{yx}\overset{!}{=} 0$$

### Initialisieren des Spannungstensors
Gegeben ist der folgenden 2D Spannungszustand:

$$\underline{\underline{\sigma}}= \begin{bmatrix} 130 & -40 \\ -40 & 70\end{bmatrix}\text{MPa}$$

Mittels des Befehls 
```python
sigma = np.array([[130, -40],[-40, 70]])
```

Wird die Variable `sigma` $\underline{\underline{\sigma}}$ im Arbeitsbereich initialisiert. <br>
Zur weiterführenden Dokumentation der Funktion `np.array` folgen Sie dem Link [NumPy-Dokumentation](https://numpy.org/doc/stable/reference/generated/numpy.array.html#numpy.array).

````python
sigma = np.array([[130, -40],[-40, 70]]);
````

### Berechnung der Eigenwerte und -vektoren 

[NumPy-Dokumentation](https://numpy.org/doc/stable/reference/generated/numpy.linalg.eig.html)


````python
eig_vals, eig_vecs = np.linalg.eig(sigma);
````

### Ergebnisse
